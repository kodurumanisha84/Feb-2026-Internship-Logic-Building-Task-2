```python
#Task-1  Unique Words in a Sentence

# Take input from user
Sentence = input("Enter sentence: ")

# Split sentence into words
words = Sentence.split()

# Convert list to set to remove duplicates
unique_words = set(words)

# Print results
print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

```python
#Task-2 Highest Salary from Employee Data

# Employee dictionary
Employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Find employee with highest salary
highest_employee = max(Employees, key=Employees.get)
highest_salary = Employees[highest_employee]

# Print result
print("Highest Salary:", highest_employee, "-", highest_salary)

```

```python
#Task-3 Find Maximum and Minimum Values
# Given list
numbers = [45, 22, 89, 10, 66]

# Initialize max and min with first element
maximum_number = numbers[0]
minimum_number = numbers[0]

# Loop through the list
for i in numbers:
    if i > maximum_number:
        maximum_number = i
    if i < minimum_number:
        minimum_number = i

# Print results
print("List:", numbers)
print("Max:", maximum_number)
print("Min:", minimum_number)
```

```python
#Task-4  Count Products Above a Price Threshold
# Given list
prices = [450, 1200, 899, 1500, 300]

count = 0

# Loop through prices
for i in prices:
    if i > 1000:
        count += 1

# Print result
print("Products above 1000:", count)
```

```python
#Task-5 Calculate Attendance Percentage
# Given list
Attendance = ["P", "P", "A", "P", "P"]

Present_count = 0

# Count number of "P"
for i in Attendance:
    if i == "P":
        Present_count += 1

# Calculate percentage
percentage = (Present_count / len(Attendance)) * 100

# Print result
print("Attendance Percentage:", percentage)
```

```python
#Task-6 Remove Duplicate Phone Numbers

# Given list
Phone_numbers = [9876543210, 9123456789, 9876543210]

# Convert list to set to remove duplicates
unique_numbers = set(Phone_numbers)

# Print result
print("Unique phone numbers:", unique_numbers)
```

```python
#Task-7  Count Character Frequency

# Given string
text = "pythonp"

# Create empty dictionary
char_count = {}

# Count each character
for i in text:
    if i in char_count:
        char_count[i] += 1
    else:
        char_count[i] = 1

# Print result
print(char_count)
```

```python
#Task-8 Convert List to Tuple

# Given list
numbers_List = [10, 20, 30]

# Convert list to tuple
numbers_tuple = tuple(numbers_List)

# Print result
print("Tuple:", numbers_tuple)
```

```python
#Task-9  Check if a Key Exists in a Dictionary

# Given dictionary
Employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Key to check
key = "Ravi"

# Check if key exists
if key in Employees:
    print("Employee exists")
else:
    print("Employee does not exist")
```
